```python
import numpy as np
import pandas as pd
import dotenv 
import requests
import json
import os
```

User-agent string


```python
r = requests.get('https://httpbin.org/user-agent')
useragent = json.loads(r.text)['user-agent']
headers = {'User-agent': useragent}
```

Load ENV

```python
dotenv.load_dotenv()
newskey = os.getenv('newskey')
```

News API

```python
params = {'apiKey' : newskey,
          'q': 'Bob Good'}
url = 'https://newsapi.org/v2/everything'
r = requests.get(url, headers = headers, params = params)
myjson = json.loads(r.text)
news_df = pd.json_normalize(myjson, record_path = ['articles'])
```

```python
news_df
```

```python
dotenv.load_dotenv()
congresskey = os.getenv('congresskey')
```

```python
root = 'https://api.congress.gov/v3'
endpoint = 'member'
params = {'api_key': congresskey, 
          'format': 'json',
          'currentMember': 'true'}
r = requests.get(root + endpoint,
                params = params,
                headers = headers)
r.text
```

```python
member_df = pd.json_normalize(json.loads(r.text))
member_df
```

```python

bobgoodinfo = member_df.query('state=="Virginia" & district==5')
bobgood_id = bobgoodinfo['bioguideId']
bobgood_id
```

```python
endpoint = f'/member/{bobgood_id}/sponsored-legislation'
```

```python
r = requests.get(root + endpoint, params = params, headers = headers)
r
```

```python
bobgood_bills = json.loads(r.text)
bobgood_bills
```

```python
[print(x) for x in bobgood_bills ['title']]
```
